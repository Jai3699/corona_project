```python
df3=pd.read_csv('covid_19_data.csv',parse_dates=['Last Update'])
```

```python
df3.rename(columns={'ObservationDate':'Date','Country/Region':'Country'},inplace=True)
```

```python
df3
```

```python
df3.query('Country=="UK"')
```

```python
#df3.groupby('Date')['Confirmed'].sum().sort_values(ascending=False).to_frame().style.background_gradient(cmap='Reds')
```

```python
confirmed=df3.groupby('Date').sum()['Confirmed'].reset_index()
# reset_index is used to make the table in dataframe with index
death=df3.groupby('Date')['Deaths'].sum().to_frame().reset_index()
recovered=df3.groupby('Date')['Recovered'].sum().reset_index()
```

```python
confirmed['Date']
```

```python
plt.scatter(x=confirmed['Date'],y=confirmed['Confirmed'],label='Confirmed',linewidth=2)
plt.scatter(death['Date'],death['Deaths'],label='Death',linewidth=2)
plt.scatter(recovered['Date'],recovered['Recovered'],label='Recovered',linewidth=2,color='blue')
plt.rcParams['figure.figsize']=(30,20)
#plt.rc('xtick',labelsize=20)
plt.legend()

```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'],y=confirmed['Confirmed'],name='Confirmed',mode='lines+markers',line=dict(color='Purple')))
fig.add_trace(go.Scatter(x=death['Date'],y=death['Deaths'],name='Deaths',mode='lines+markers'))
fig.add_trace(go.Scatter(x=recovered['Date'],y=recovered['Recovered'],name='Recovered',mode='lines+markers'))
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=confirmed['Date'],y=confirmed['Confirmed'],name='Confirmed',width=3))
fig.add_trace(go.Bar(x=death['Date'],y=death['Deaths'],width=3,name='Deaths'))
fig.add_trace(go.Bar(x=recovered['Date'],y=recovered['Recovered'],width=3,name='Recovered'))
```
